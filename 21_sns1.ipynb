**Seaborn Vs Matplotlib**

*It is summarized that if Matplotlib “tries to make easy things easy and hard things possible”, Seaborn tries to make a well-defined set of hard things easy too.”*

*Seaborn helps resolve the two major problems faced by Matplotlib; the problems are −*

1.   Default Matplotlib parameters
2.   Working with data frames

# Plotting using Seaborn

**Histogram using distance plotting**
- displot
- distplot

Histograms represent the data distribution by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.

```python
#Flexibly plot a univariate distribution of observations.
sns.displot(t.age)
```

```python
# will use DataFrame
# if we want to change default name on X axis then
# use matplotlib
sns.displot(data=t,x='age')
plt.xlabel('Age Data')
```

```python
# add bins
sns.displot(data=t,x='age',bins=[10,20,30,40])
plt.xlabel('Age Data')
```

```python
#change color palette
sns.displot(data=t,x='age',color='red')
```

```python
#combine 2 columns and take a hist plot
sns.displot(t[['age','fare']])
```

**Plotting Bivariate Distribution: Joint Plot**

Bivariate Distribution is used to determine the relation between two variables. 

```python
#It gives scatter plot+ hist plot & its strictly bivariate plot
sns.jointplot(x='species',y='flipper_length_mm',data=p)
```

```python
i = sns.load_dataset('iris')
sns.jointplot(x='species',y='sepal_length',data=i)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=i)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=i,hue='species')
```

#### kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=i,hue='species',kind='kde')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=i,kind='hex')
```

```python
#use regression plot
sns.jointplot(x='sepal_length',y='sepal_width',data=i,kind='reg')
```

**Pair relationships using pairplot**

To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function.

```python
#pairplot works on Numeric data only
# matrix plot which display column to column relationship
sns.pairplot(i)
```

```python
#hue is used on categorical data
sns.pairplot(i,hue='species')
```

```python
# pair plot for sepal_length and petal_length
le = i.loc[:,['sepal_length','petal_length','species']]
sns.pairplot(le,hue='species',palette='gist_earth')
```
